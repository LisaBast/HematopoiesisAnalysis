```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

# Mann Whitney U test to test significant differences between healthy and MDS in Bone Marrow (BM) snapshot data

```python
df_FACS_snapshot = pd.read_csv('./hematopoiesis_sample_snapshot.txt', sep="\t", na_values=['NaN'])
```

```python
df_FACS_snapshot = df_FACS_snapshot.drop('Unnamed: 7',axis=1).copy()
D = df_FACS_snapshot[:-2]
D = D.apply(lambda x: x.str.replace(',','.'))
D
```

```python
np.mean(D[D['group']=='H']['HSC'].astype(float).to_numpy())
```

```python
len(D.columns)
```

```python
for ct_id, CT in enumerate(D.columns[1:len(D.columns)]):
    S_H = D[D['group']=='H'][CT].astype(float).to_numpy()
    S_MDS = D[D['group']=='MDS'][CT].astype(float).to_numpy()
    M_H = np.mean(S_H)
    M_MDS = np.mean(S_MDS)
    if M_H<M_MDS:            
        opt_hyp = 'less'
    else:
        opt_hyp = 'greater'
    TR_MWU = stats.mannwhitneyu(S_H, S_MDS, alternative=opt_hyp) #0 and 1
    print(CT+' p_val for H1: '+opt_hyp+': '+str(TR_MWU[1]))
    print(CT+' corrected p_val for H1: '+opt_hyp+': '+str(TR_MWU[1]*(len(D.columns)-1)))
    
```
